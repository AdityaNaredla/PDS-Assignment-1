```python

raw_data = "/content/drive/MyDrive/PDS/Q1/data/raw/raw_fraility_data.csv"
from sklearn.preprocessing import StandardScaler
raw_data = "/content/drive/MyDrive/PDS/Q1/data/raw/raw_fraility_data.csv"
df = pd.read_csv(raw_data)
print("Initial Data:")
print(df.head())
df['Frailty'] = df['Frailty'].map({'N': 0, 'Y': 1})
numeric_columns = ['Height', 'Weight', 'Age', 'Grip strength']
scaler = StandardScaler()
df[numeric_columns] = scaler.fit_transform(df[numeric_columns])
print("Processed Data:")
print(df.head())
```

```python
processed_data = "/content/drive/MyDrive/PDS/Q1/data/processed/processed_fraility_data.csv"
df.to_csv(processed_data, index=False)
```

```python
plt.figure(figsize=(12, 10))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='RdYlGn', fmt=".2f")
plt.title('Correlation Matrix of Frailty Data')
plot_path = "/content/drive/MyDrive/PDS/Q1/results/correlation_matrix.png"
plt.savefig(plot_path)
plt.show()

```
